```python
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

```python
df = pd.read_csv("/Users/Joan/Desktop/Capstone/sugar_consumption.csv", index_col = 'NA.')
df.head()
```

```python
data1961 = df[['X1961']]
data1961.head(5)
```

```python
data1961.plot()
plt.show()
```

```python
condensed_data1961 = data1961.dropna()
condensed_data1961.head(15).plot.bar()
plt.title("Sugar Consumption in 1961 by Country")
plt.xlabel("Country Index")
plt.ylabel("Sugar Consumption in grams/day")
plt.show()
```

```python
avg_sugar_consumption_1961 = df.groupby('NA.').mean()['X1961'].dropna()
avg_sugar_consumption_1961
```

```python
avg_1961_ranked = avg_sugar_consumption_1961.sort_values(ascending=False).head(15)
avg_1961_ranked
```

```python
avg_1961_ranked.plot.bar(x='country', y='x1961')
plt.title("Average Sugar Consumption by Country in 1961")
plt.ylabel("Average Sugar Consumption in grams/day")
```
